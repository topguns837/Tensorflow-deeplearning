```python
#Create features
X = np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0])

#Create labels
y = np.array([3.0,6.0,9.0,12.0,15.0,18.0,21.0,24.0])
```

 ### Steps in Modelling Tensorflow

```python
# 1.Creating a model
# 2.Compiling a model
# 3.Fitting a model
```

```python
# Set random seed
tf.random.set_seed(42)

#1. Create a model using Sequential API
model = tf.keras.Sequential([
   tf.keras.layers.Dense(1)                          
])

#2. Compile the model
model.compile(loss = tf.keras.losses.mae,          #mae = mean absolute error
              optimizer = tf.keras.optimizers.SGD(),  #SGD = schostic gradient descent
              metrics = ["mae"]) 

#3. Fit the model

model.fit(tf.expand_dims(X, axis=-1), y, epochs=5)
```

```python
# Check out X and y
X,y
```

```python
# Try and make a prediction using our model
model.predict([17.0])
```

### Improving our model

We can improve our model ny altering the steps we took to create a model.

1) **Creating a model** : We can add more layers, increase the no of neurons, change activation function

2) **Compiling a model** : We can change the optimization function or the learning rate of the optimization function

3) **Fitting a model** : We can train for more epochs, and increase training size

```python
# Rebuilding the model with more epochs

# Set random seed
tf.random.set_seed(42)

#1. Create a model using Sequential API
model = tf.keras.Sequential([
   tf.keras.layers.Dense(1)                          
])

#2. Compile the model
model.compile(loss = tf.keras.losses.mae,          #mae = mean absolute error
              optimizer = tf.keras.optimizers.SGD(),  #SGD = schostic gradient descent
              metrics = ["mae"]) 

#3. Fit the model

model.fit(tf.expand_dims(X, axis=-1), y, epochs=1000)


```

```python
#Let's see if model predictions also improve
model.predict([17])
```

```python
# Changing the optimizer to Adam

#1. Create a model

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1)
])

#2 Compile the model

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(learning_rate=0.0001),
              metrics = ["mae"])

#3. Fit the model
model.fit(tf.expand_dims(X,axis=1),y,epochs=100)

```

```python
# Using ReLu activation

model = tf.keras.Sequential([
                             tf.keras.layers.Dense(50,activation="relu"),
                             tf.keras.layers.Dense(1)
])

#2. Compile the model

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics = ["mae"])

#3.Model  fitting
model.fit(tf.expand_dims(X, axis=-1), y, epochs=1000)
```

```python
model.predict([17.0])
```

### Splitting data into train and test set

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8, random_state=42)
# sklearn train_test_evaluation doesnt work with tensors
```

```python
tf.random.set_seed(42)

X_train,X_test = tf.split(X,[40,10], axis=0)
```

```python
X_train ,  X_test
```

```python
y_train,y_test = tf.split(y,[40,10], axis=0)
```

```python
y_train,y_test
```

### Data Visualization

```python
plt.figure(figsize = (10,7))

#Plot training data in blue
plt.scatter(X_train, y_train, c="b", label = "Training Data")

#Plot the test data in green
plt.scatter(X_test,y_test,c="r",label = "Test Data")

# Show a legend also
plt.legend()



```

```python
# Let's have a look at out neural network

#1. Create a model
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(1,input_shape = [1,]),
                             tf.keras.layers.Dense(1,name="output_layer")
],name="deep_learning")

#2. Compile the model

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

```

```python
# Visualizing the model
model.build()
model.summary()
```

```python
# Total params = Total params in the model
# Trainable params = No. of params the model can update during training
# Non-trainable params = These params aren;t updated during training ( Used in transfer learning) 
```

```python
model.weights
```

```python
#3. Fit the model
model.fit(X_train,y_train,epochs = 100,verbose = 0)
```

```python
y_pred = model.predict(X_test)
```
